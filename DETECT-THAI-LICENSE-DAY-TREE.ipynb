```python
# -*- coding: utf-8 -*-

# http://doc.openalpr.com/cloud_api.html
# https://cloud.openalpr.com/cloudapi/
import requests
import base64
import json
from sys import argv
import cv2
import matplotlib.pyplot as plt


IMAGE_PATH = "C:/Users/NRRU/Desktop/python/w644.jpg"
LABEL_PATH = IMAGE_PATH.replace("/img/","/label/")
SECRET_KEY = 'sk_cbf88d6ea3e66e673736b76a'

with open(IMAGE_PATH, 'rb') as image_file:
    img_base64 = base64.b64encode(image_file.read())

url = 'https://api.openalpr.com/v2/recognize_bytes?recognize_vehicle=1&country=th&secret_key=%s' % (SECRET_KEY)
r = requests.post(url, data = img_base64)

t = json.dumps(r.json(), indent=2)
#print(t)
y = json.loads(t)
img = cv2.imread(IMAGE_PATH)
plt.imshow(img);plt.show()
```

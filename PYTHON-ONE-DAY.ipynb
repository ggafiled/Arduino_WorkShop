```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:/Users/NRRU/Desktop/python/tom.jpg")    

matrix = np.ones(img.shape,dtype="uint8")*100
imadd = cv2.add(img,matrix)
matrix = np.ones(img.shape,dtype="uint8")*50
imsubtract = cv2.subtract(img,matrix)

_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

haarcascade = cv2.CascadeClassifier('C:/Users/NRRU/Desktop/python/haarcascades/haarcascade_frontalface_default.xml')
faceDetech = haarcascade.detectMultiScale(imgray,scaleFactor=1.5,minNeighbors=5)
for(x,y,w,h) in faceDetech:
    cv2.rectangle(_img, (x, y), (x+w, y+h), (255,0, 0), 2)

plt.subplot(2,2,1)
plt.imshow(img)
plt.subplot(2,2,2)
plt.imshow(imgray)

plt.figure()
plt.imshow(_img)
 
```

```python
video_capture = cv2.VideoCapture(0)
while True:
    ret, frame = video_capture.read()
    imgray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    haarcascade = cv2.CascadeClassifier('C:/Users/NRRU/Desktop/python/haarcascades/haarcascade_frontalface_default.xml')
    faceDetech = haarcascade.detectMultiScale(imgray,scaleFactor=1.2,minNeighbors=5)
    for(x,y,w,h) in faceDetech:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255,0, 0), 2)
        cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
